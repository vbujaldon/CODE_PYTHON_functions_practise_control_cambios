#### 1.1 Calculadora IMC (√çndex de massa corporal)
Escriu una funci√≥ que calculi l'IMC ingressat per l'usuari/√†ria, √©s a dir, qui ho executi haur√† d'ingressar aquestes dades.
La funci√≥ ha de classificar el resultat en les seves respectives categories.
Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introdu√Ødes per l‚Äôusuari/a no estan en el format adequat o no pren valors raonables.

```python
def pedir_medida(mensaje_medida):
    while True:
        try:
            return float(input(mensaje_medida))
        except ValueError:
            print(("Error: debes escribir la medida con un n√∫mero y un punto, si deseas poner decimales. Vuelve a probar."))

height_user = pedir_medida("Introduce tu altura en metros. Por ejemplo '1.70'.")
weight_user = pedir_medida("Introduce tu peso en kg. Por ejemplo: '66.5'.")

def calculo_imc(weight_user, height_user):
    return weight_user / (height_user**2)

imc_calculado = calculo_imc(weight_user, height_user)

if imc_calculado > 30:
    mensaje = "Tienes obesidad, contacta a tu m√©dico."
elif imc_calculado >= 25:
    mensaje = "Tienes sobrepeso, consulta con tu m√©dico."
elif imc_calculado >= 18.5:
    mensaje = "Tienes un peso normal. Enhorabuena."
else:
    mensaje = "Tienes un peso demasiado bajo. Contacta a tu m√©dico."

print(f'Tu IMC (√≠ndice de masa corporal) es {imc_calculado: .2f}. {mensaje}')
```

#### 1.2. Conversor temperatures
Selecciona 2 conversors, de tal manera que en introduir una temperatura retorni, com a m√≠nim, dues conversions, i de manera que es puguin guardar (recorda que un print() no es pot guardar mai).
Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introduides per l‚Äôusuari/a no estan en el format adequat.
(EXTRA): Pensa una manera d‚Äôemmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d‚Äôescriure molts if else en funci√≥ de la temperatura d‚Äôorigen i la temperatura de dest√≠.
Nota: lambda serveix per crear funcions curtes d'una l√≠nia

```python
# Diccionario conversiones para evitar multiples if/elif
dicc_conversiones = {
    "Celsius" : {
        "Farenheit": lambda c: c * 9/5 +32,
        "Kelvin": lambda c: c+273.15
    },
    "Farenheit": {
        "Celsius": lambda f: (f-32) * 5/9,
        "Kelvin": lambda f: (f-32) * 5/9 +273.15
    },
    "Kelvin": {
        "Celsius": lambda k: k-273.15,
        "Farenheit": lambda k: (k-273.15) * (9/5) +32
    }
}

# funci√≥n validar entrada T¬™(num) user
def pedir_validar_temp(mensaje):
    while True:
        try:
            return float(input(mensaje))
        except ValueError:
            print(("Error: debes escribir la temperatura con un n√∫mero (y un punto, si deseas poner decimales). Vuelve a probar."))

# funci√≥n validar entrada user (unidades acotadas)
def pedir_validar_unidad(mensaje, lista_unidades):
    while True:
        unidad = input(mensaje).capitalize()
        if unidad in lista_unidades:
            return unidad
        else:   
            print(f"Error: debes elegir la unidad de medida de la temperatura " f"de entre las siguientes: {', '.join(lista_unidades)}")

# lista para delimitar entrada unidad temperatura
lista_unidades = list(dicc_conversiones.keys())

# llamada funci√≥n validar/input + guardar valores
temp_usuario = pedir_validar_temp("Introduce la temperatura. Por ejemplo '36.70'.")
unidad_usuario = pedir_validar_unidad("Introduce la medida de origen (Kelvin, Celsius o Farenheit).", lista_unidades)

# Aplicar conversi√≥n unidades y guardar resultado en nuevo dicc
dicc_resultados = {}
for unidad_final, funcion in dicc_conversiones[unidad_usuario].items():
    dicc_resultados[unidad_final] = funcion(temp_usuario)

# Mostrar resultados
print(f"Temperatura original: {temp_usuario}¬∞ {unidad_usuario}, que equivalen a: \n ")
for unidad_final, valor_convertido in dicc_resultados.items():
    print(f"{unidad_final}: {valor_convertido:.2f}¬∞{unidad_final[0]}")
```

#### 1.3. Comptador de paraules d'un text.

Escriu una funci√≥ que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casu√≠stiques possibles que facin que el programa no funcioni correctament.
(EXTRA): Quina √©s la llargada mitjana de les paraules del text que has escrit? ‚ÄúHola com va?‚Äù hauria de retornar (4+3+2) / 3 = 3

```python
def recuento_palabras(texto):
    dic_palabras = {}
    for simbol in ".,;:!?¬ø¬°()": # elimina s√≠mbolos x evitar palabras duplicadas (Hola = Hola!)
        texto = texto.replace(simbol, "")

    for palabra in texto.split():
        palabra_minuscula = palabra.lower()
        if palabra_minuscula in dic_palabras:
            dic_palabras[palabra_minuscula] += 1
        else:
            dic_palabras[palabra_minuscula] = 1
    return dic_palabras

texto = input("Introduce el texto que quieres analizar")
resultado_recuento = recuento_palabras(texto)

print(f' Recuento de palabras en el texto: \n {resultado_recuento}')


# EXERCICIO EXTRA: c√°lculo promedio longitud de las palabras
def promedio_long_palabras(texto):
    for simbol in ".;:,!¬°¬ø?":
        texto = texto.replace(simbol, "")

    palabras = texto.split() #lista con palabras separadas

     #if len(palabras) == 0: return 0

    total_letras = 0
    for palabra in palabras:
        total_letras += len(palabra)

    return round(total_letras / len(palabras),2)

longitud_texto = promedio_long_palabras(texto)
print(f"El promedio de la longitud de las palabras del texto:' {longitud_texto}")
```

#### 1.4. Invertir un diccionari
Intercanvia les claus i els valors d'un diccionari. Els valors i claus en el diccionari original haurien de ser √∫nics, per√≤ si no √©s el cas, la funci√≥ ha d'imprimir un missatge d'advert√®ncia, juntament amb una llista amb els valors associats a la clau repetida.

```python
def inversion_diccionario(dic_original):
    dic_invertido = {}
    dic_llaves_duplicadas = {}
    for key, value in dic_original.items():
        if value not in dic_invertido:
            dic_invertido[value] = key
        else:   
            if value not in dic_llaves_duplicadas:
                dic_llaves_duplicadas[value] = [dic_invertido[value]]
            dic_llaves_duplicadas[value].append(key)
    if dic_llaves_duplicadas:
        print("ALERTA: hay valores duplicados en el diccionario original que impiden crear todos los registros en el diccionario invertido:")
        for value, key in dic_llaves_duplicadas.items():
            print(f'Valor {value} repetidos en las siguientes claves: {key}')
    return dic_invertido

diccionario_fruta = {"x": "apple", "y":"banana", "z": "banana"}
inversion_diccionario(diccionario_fruta)
```

#### 2.1 Comptador i endre√ßador de paraules d'un text
El client va quedar content amb el comptador de paraules, per√≤ ara vol llegir arxius TXT i que calculi la freq√º√®ncia de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, √©s a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
# Subir fichero y transformar a string:
def subir_file_pasar_txt_str(link_carpeta):
    with open(link_carpeta, "r") as fichero: #para cierre fichero autom
        texto_a_analizar = fichero.read()
    return texto_a_analizar

def recuento_palabras(texto):
    dic_palabras = {}
    for simbol in ".,;:!?¬ø¬°()[]{}-<>...&*'\"": # elimina s√≠mbolos x evitar palabras duplicadas (Hola = Hola!)
        texto = texto.replace(simbol, "")

    for palabra in texto.split():
        palabra_minuscula = palabra.lower()
        if palabra_minuscula in dic_palabras:
            dic_palabras[palabra_minuscula] += 1
        else:
            dic_palabras[palabra_minuscula] = 1
    return dic_palabras

# Ordenar palabras y transformar a diccionario
def ordenar_palabras_int_ext(diccionario):
    diccionario_ordenado = {}
    for primera_letra, subdic in sorted(diccionario.items()):
        diccionario_ordenado[primera_letra] = dict(sorted(subdic.items())) #dict= crear dic y no lista de tuplas
    return diccionario_ordenado

# crear dicc dentro de dic con formato {1a letra: {palabras:recuento}}
def agrupar_1a_letra(diccionario):
    dic_agrupado_1a_letra = {}
    for palabra_clave, valor in diccionario.items():
        primera_letra = palabra_clave[0]
        if primera_letra not in dic_agrupado_1a_letra:
            dic_agrupado_1a_letra[primera_letra] = {}
        dic_agrupado_1a_letra[primera_letra][palabra_clave] = valor
    return dic_agrupado_1a_letra



# Ejecutamos subida archivo y conversi√≥n a string
texto_blanca = subir_file_pasar_txt_str(r'C:\Users\vanes\OneDrive\Documenten\CURSOS\IT Academy - An√†lisis de dades\_ESPECIALITAT\Sprint 9 Python\tu me quieres blanca.txt')

# Ejecutamos recuento palabras (crea dic con "palabra:repeticiones")
dic_con_recuento = recuento_palabras(texto_blanca)

# Ejecutamos crear dic con formato 1letra + subdic:
diccionario_agrupado = agrupar_1a_letra(dic_con_recuento)

# Ejecutamos ordenaci√≥n diccionario
dic_ordenado = ordenar_palabras_int_ext(diccionario_agrupado)

# Print en vertical:
for clave_externa, subdiccionario in dic_ordenado.items(): #items() para mostrar tmb value
    print(f"{clave_externa}:")
    for clave_interna, valor in subdiccionario.items():
        print(f" {clave_interna}: {valor}")

```

#### 2.2 Conversi√≥ de tipus de dades
El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python
# Transformar lista variada con tuplas i listas a lista simple:
def simplificar_lista_mixta_con_iterables(lista_externa):
    lista_simplificada = []
    for elemento in lista_externa:
        if isinstance(elemento, (list, tuple)): # pregunta si elemento = lista o tuple -> True
            lista_simplificada.extend(simplificar_lista_mixta_con_iterables(elemento)) #se llama otra vez a la funci√≥n para extraer y clasificar elemento individual en lista / .extend para poner varios elementos en una lista (.append solo 1 elemento)
        else:
            lista_simplificada.append(elemento)
    return lista_simplificada

# Separar elementos seg√∫n si son convertibles a float:
def creacion_listas_float_no_float(lista_externa):
    elem_float_ok = []
    elem_no_float = []
    for objeto in lista_externa:
        try: # se usa para ejecutar c√≥digo que puede fallar
            elem_float_ok.append(float(objeto))
        except (ValueError, TypeError):
            elem_no_float.append(objeto)
    return elem_no_float, elem_float_ok

lista_ready_para_float = simplificar_lista_mixta_con_iterables(['1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']])
lista_elem_no_float, lista_float_ok = creacion_listas_float_no_float(lista_ready_para_float)
print("Lista de elementos convertidos a float:", lista_float_ok)
print("Lista de elementos NO convertibles a float:", lista_elem_no_float)
```

#### 3.1 Generador de Contrasenyes
Explora el funcionament del m√≤dul random de la llibreria numpy. üîó Random module in NumPy

En aquest punt, el client ens ha encarregat una funci√≥ de Python que generi contrasenyes segures. La funci√≥ ha de dependre dels seg√ºents par√†metres:
longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d‚Äôapar√®ixer maj√∫scules
minuscules (bool = True): Si hi ha d‚Äôapar√®ixer min√∫scules
numeros (bool = True): Si hi ha d‚Äôapar√®ixer n√∫meros
signes (bool = False): Si hi ha d‚Äôapar√®ixer car√†cters especials (,-$? o similars)
Aix√≠ doncs, si executem la funci√≥ de la seg√ºent manera:

crear_contrasenya(10, True, True, True, True)

Haur√≠em d‚Äôobtenir un output (que hem de poder guardar) de l‚Äôestil:

9Er,5Vn8P$

Assegura‚Äôt que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleat√≤ries.

(EXTRA) Explora com podr√≠em fer que la funci√≥ copi√©s la contrasenya autom√†ticament al porta papers de l‚Äôordinador (com si l‚Äôhagu√©ssim seleccionat i fet ctrl+copy).


```python
import numpy as np
import string
import pyperclip

# Funci√≥n crear contrse√±a
def crear_password(longitud, majusc, minusc, numeros, signes):
    lista_caracteres = "" # crea cadena vac√≠a
    
    if majusc: # comprueba si True (hay)
        lista_caracteres = lista_caracteres + string.ascii_uppercase
    if minusc:
        lista_caracteres = lista_caracteres + string.ascii_lowercase
    if numeros:
        lista_caracteres = lista_caracteres + string.digits
    if signes:
        lista_caracteres = lista_caracteres + string.punctuation
    if not lista_caracteres:
        raise ValueError("Selecciona m√≠nimo un tipo de caracteres para generar la contrase√±a")
    
    password = ''.join(np.random.choice(list(lista_caracteres), size=longitud))
    return password

# Ejecutamos crear contrase√±a
password_generado = crear_password(10, True, True, True, False)
print(password_generado)

# EXTRA- funci√≥n para copiar autom√°ticamente el text
pyperclip.copy(password_generado)
```

#### 3.2 Processament de dades simple

Fitxer amb hist√≤ric de partits de futbol catal√†, amb noms dels equips i els resultats. Necessita que processem les dades de manera autom√†tica, per extreure‚Äôn els resultats que necessita. Utilitza el fitxer "historic_partits.txt"

El programa ha de retornar:

- El nombre total de gols que ha fet cada equip.
- El nom de l‚Äôequip m√©s golejador.
- El nom de l‚Äôequip m√©s golejat
- La classificaci√≥ global (cada vict√≤ria: 3 pts, empat 1 pts, derrota 0 pts)

```python
# Obrim el fitxer
with open(r"C:\Users\vanes\OneDrive\Documenten\CURSOS\IT Academy - An√†lisis de dades\_ESPECIALITAT\Sprint 9 Python\historic_partits.txt", "r") as fichero:
    lines = fichero.readlines()

# Diccionaris per guardar dades
gols_fets = {}
gols_rebuts = {}
punts = {}

# Processament de cada l√≠nia
for linea in lines:
    linea = linea.strip()
    if linea == "":
        continue  # saltar l√≠nies buides
    dades = linea.split("\t")
    if len(dades) != 3:
        continue  # saltar l√≠nies mal formatades

    equip_local, resultat, equip_visitant = dades
    gols_local, gols_visitant = map(int, resultat.split("-"))

    # Actualitzar gols fets
    gols_fets[equip_local] = gols_fets.get(equip_local, 0) + gols_local
    gols_fets[equip_visitant] = gols_fets.get(equip_visitant, 0) + gols_visitant

    # Actualitzar gols rebuts
    gols_rebuts[equip_local] = gols_rebuts.get(equip_local, 0) + gols_visitant
    gols_rebuts[equip_visitant] = gols_rebuts.get(equip_visitant, 0) + gols_local

    # Actualitzar punts
    if gols_local > gols_visitant:
        punts[equip_local] = punts.get(equip_local, 0) + 3
        punts[equip_visitant] = punts.get(equip_visitant, 0) + 0
    elif gols_local < gols_visitant:
        punts[equip_local] = punts.get(equip_local, 0) + 0
        punts[equip_visitant] = punts.get(equip_visitant, 0) + 3
    else:  # empat
        punts[equip_local] = punts.get(equip_local, 0) + 1
        punts[equip_visitant] = punts.get(equip_visitant, 0) + 1

# Resultats
equip_mes_golejador = max(gols_fets, key=gols_fets.get)
equip_mes_golejat = max(gols_rebuts, key=gols_rebuts.get)
classificacio = sorted(punts.items(), key=lambda x: x[1], reverse=True)

# Mostrar resultats
print("Gols fets per equip:")
for equip, gols in gols_fets.items():
    print(f"{equip}: {gols}")

print("\nEquip m√©s golejador:", equip_mes_golejador)
print("Equip m√©s golejat:", equip_mes_golejat)

print("\nClassificaci√≥ global (Punts):")
for equip, punt in classificacio:
    print(f"{equip}: {punt}")
```
